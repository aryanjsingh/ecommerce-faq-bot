## 1. Written Summary

**Domain:** E-Commerce Customer Support
**User:** Online Shoppers
**What the agent does:** Handles common queries from the product catalogue and return policies without hallucinating.
**KB Size:** 10 Document Chunks
**Tool used:** `get_current_time` to answer time-related questions dynamically.

**Baseline Testing Results:**
- Route tests: Passed 10/10
- Faithfulness scores: 1.0 on all retrieved responses
- Out of scope: Passed (Refused strictly)
- Sliding window: Passed (Remembered user name and previous query context over 3+ turns)

## 2. One thing I would improve with more time

With more time, I would improve the Vector Retrieval by introducing **Hybrid Search** (combining sparse Keyword Search like BM25 with dense Embeddings). Currently, semantic similarity alone can fail on exact product ID queries or very technical keyword matching (like 'SKU-00123'). Adding an ensemble retriever would increase precision when users request specific standard policies versus targeted keywords.

## 3. Component Test

```python
from agent import setup_kb, create_graph, get_llm

llm = get_llm()
embedder, collection = setup_kb()
app = create_graph(llm, embedder, collection)
print("Graph compiled successfully!")
```

```python
import uuid
def ask(question, thread_id):
    config = {"configurable": {"thread_id": thread_id}}
    result = app.invoke({"question": question}, config=config)
    print("\n===========================")
    print(f"Question: {question}")
    print(f"Route Used: {result.get('route')}")
    print(f"Faithfulness Score: {result.get('faithfulness')}")
    print(f"Answer:\n{result.get('answer')}")
    print("===========================")
```

```python
# Multi-turn context test
mock_session_id = str(uuid.uuid4())
ask("Hi, my name is Alex! What is your return policy?", mock_session_id)
ask("What if it is a gift card?", mock_session_id)
ask("Since you know my name, are gift cards refundable?", mock_session_id)
ask("What is the current time?", mock_session_id)
ask("Do you sell prescription medicine?", mock_session_id) # Out of scope test
```
